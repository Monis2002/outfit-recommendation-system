# Random date and time

```python
import random
from datetime import datetime, timedelta

def generate_random_datetime():
    # Define the start and end dates
    start_date = datetime(2000, 1, 1)
    end_date = datetime(2020, 12, 31)

    # Generate a random number of days between start and end
    random_days = random.randint(0, (end_date - start_date).days)

    # Calculate the random date
    random_date = start_date + timedelta(days=random_days)

    # Generate random hours, minutes, and seconds
    random_hours = random.randint(0, 23)
    random_minutes = random.randint(0, 59)
    random_seconds = random.randint(0, 59)

    # Combine date and time components
    random_datetime = random_date.replace(
        hour=random_hours, minute=random_minutes, second=random_seconds
    )

    return random_datetime

# Example usage
random_datetime = generate_random_datetime()
print("Random Date and Time:", random_datetime)
```

```python
dic_list=[]
path_list=os.listdir('dataset\\formal_male_winter')
weather=input('Weather  ')
Type=input("Type  ")

for path in path_list:
    r=os.path.join('dataset\\formal_male_winter',path)
    d={'Date':generate_random_datetime(),'Type':Type,'Gender':"M",'weather':weather,'Path':r}
    dic_list.append(d)
```
